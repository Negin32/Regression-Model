# **Imputing Outliers in numerical variables**

```python
#Imputes outliers by replacing them with the median of the column.
#Uses the Interquartile Range (IQR) method to detect outliers.
#Steps:
#Identifies numerical columns.
#Calculates Q1 (25th percentile) and Q3 (75th percentile).
#Computes the IQR (Q3 - Q1).
#Defines lower and upper bounds using 1.5 Ã— IQR.
#Detects outliers based on these bounds.
#Replaces outliers with the median of the respective column.


def impute_outliers(df):
    numerical_columns = df.select_dtypes(include=np.number).columns # Identify numerical columns
    for column in numerical_columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identify outliers
        outliers = (df[column] < lower_bound) | (df[column] > upper_bound)

        # Replace outliers with the median of the column
        df.loc[outliers, column] = df[column].median()

    return df

# Calls the impute_outliers function to remove outliers from df.
df_imputed = impute_outliers(df)

# Display the DataFrame after imputation
print(df_imputed)
```

# **4.One Hot Encoding on Target Variable**

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['diagnosis']=le.fit_transform(df['diagnosis'])

#Encoding Categorical Variables (Diagnosis Column).The diagnosis column contains categorical values (Malignant, Benign).
#LabelEncoder converts:
#Malignant (M) â†’ 1
#Benign (B) â†’ 0
#value_counts() checks the distribution of classes.
```

# **5.Declare feature vector and target variable**

```python
X=df.drop('diagnosis',axis=1) # Features ,All columns except diagnosis.
y=df['diagnosis'] # Target Variable ,The diagnosis column (0 for Benign, 1 for Malignant).
```

# **6.Split data into separate training and test set**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
#train_test_split function is used to split the dataset into:
#Training Set (67%) â†’ Used to train the model.
#Testing Set (33%) â†’ Used to evaluate model performance.
#random_state=42 ensures reproducibility (same split every time).
```

# **8.Feature Scaling**

```python
X_train.describe() #Summarizes the numerical data (mean, median, standard deviation, min, max).
```

```python
cols=X_train.columns #Stores the column names of X_train into a variable cols.X_train.columns returns a list-like object containing the names of all columns in X_train.
#Why is this needed?
#After applying StandardScaler, X_train is transformed into a NumPy array (which loses column names).
#Later, when converting the standardized X_train back to a DataFrame, we need the original column names.
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train = pd.DataFrame(X_train, columns=[cols])
X_test = pd.DataFrame(X_test, columns=[cols])

#Standardization ensures that all features have zero mean and unit variance.
#Why standardization?
#Many ML models perform better when features are on the same scale.
#Prevents some features from dominating others due to different scales.
```

```python
X_train.describe()
```

# **9.Model training**

```python
from sklearn.linear_model import LogisticRegression
mymodel=LogisticRegression(solver='liblinear')
mymodel.fit(X_train,y_train)

#This initializes and trains a Logistic Regression model using the training dataset.
#Logistic Regression is a common algorithm for binary classification (in this case, predicting whether a tumor is benign (0) or malignant (1)).
#solver='liblinear' is used for small datasets where L1 (Lasso) or L2 (Ridge) regularization is needed.
```

# **10.Predict results**

```python
y_pred_test = mymodel.predict(X_test)
y_pred_test
#The trained model predicts the diagnosis for the test dataset (X_test).
#y_pred_test stores the predicted class labels (0 for benign, 1 for malignant).
```

# **11.Check accuracy score**

```python
from sklearn.metrics import accuracy_score
print('Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_pred_test)))
#accuracy_score(y_test, y_pred_test) compares the predicted labels (y_pred_test) with the actual labels (y_test).This means the model correctly predicts 97% of test cases.
```

# **Compare the train-set and test-set accuracy**

```python
y_pred_train = mymodel.predict(X_train)
y_pred_train
#Why Compare Train and Test Accuracy?
#Helps detect overfitting and underfitting:
#If Train Accuracy â‰« Test Accuracy â†’ Overfitting (model memorized training data, performs poorly on new data).
#If Train Accuracy â‰ˆ Test Accuracy â†’ Good Fit (balanced model).
#If Both are Very Low â†’ Underfitting (model is too simple, not capturing patterns).
```

```python
print('Training-set accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_pred_train))) #No major gap â†’ Model is well-balanced.
```

# **Check for overfitting and underfitting**

```python
print('Training set score: {:.4f}'.format(mymodel.score(X_train, y_train)))
print('Test set score: {:.4f}'.format(mymodel.score(X_test, y_test)))
#mymodel.score(X_train, y_train) returns training accuracy.
#mymodel.score(X_test, y_test) returns test accuracy.
#If training score is much higher than the test score, the model is overfitting.
#If both scores are low, the model is underfitting.
```



The training-set accuracy score is 0.9711 while the test-set accuracy to be 0.9787. These two values are quite comparable. So, there is no question of overfitting.

# **13.Classification metrices**

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_test))

#Generates a full classification report with:
#Precision: How many positive predictions were actually correct?
#Recall (Sensitivity): How many actual positives were correctly identified?
#F1-Score: Harmonic mean of Precision & Recall.
#Support: Number of actual occurrences of each class.
```

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_test , pos_label=1)
plt.plot(fpr, tpr, linewidth=2)
plt.plot([0, 1], [0, 1], 'k--' )
plt.rcParams['font.size'] = 12
plt.title('ROC curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()

#ROC Curve plots True Positive Rate (TPR) vs. False Positive Rate (FPR).
#Ideal Model â†’ Curve close to top-left corner.
#Diagonal Line (y = x) represents random guessing.
```

# **14.ROC - AUC**

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred_test)

#Measures overall model performance.
#AUC Score Meaning:
#0.5 â†’ Random guessing.
#0.7 - 0.8 â†’ Good.
#0.8 - 0.9 â†’ Very good.
#0.9 - 1.0 â†’ Excellent.
```
